LINEAR SEARCH

```python
def linear_search(log_array, target_error):
    """Linearly inspects every index for the target string."""
    for index in range(len(log_array)):
        if log_array[index] == target_error:
            return index  # Target found at this database index
    return -1  # Target not found
```

BINARY SEARCH

```python
def binary_search(sorted_ids, target_id):
    """Repeatedly divides the sorted search interval in half."""
    low = 0
    high = len(sorted_ids) - 1

    while low <= high:
        mid = (low + high) // 2
        if sorted_ids[mid] == target_id:
            return mid  # Instantly found the index
        elif sorted_ids[mid] < target_id:
            low = mid + 1  # Discard left half
        else:
            high = mid - 1  # Discard right half
    return -1
```
