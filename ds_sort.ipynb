BUBBLE SORT

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        # Last i elements are already in place
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                # Swap adjacent elements
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
```

MERGE SORT

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
        
    mid = len(arr) // 2
    left_half = merge_sort(arr[:mid])
    right_half = merge_sort(arr[mid:])
    
    return merge(left_half, right_half)

def merge(left, right):
    sorted_result = []
    i = j = 0
    
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            sorted_result.append(left[i])
            i += 1
        else:
            sorted_result.append(right[j])
            j += 1
            
    sorted_result.extend(left[i:])
    sorted_result.extend(right[j:])
    return sorted_result
```

```python
import random
import time
import copy

# =========================================================================
# 1. BUBBLE SORT (O(n²) - Brute-Force Sweeper)
# =========================================================================
def bubble_sort(arr):
    """
    Repeatedly steps through the list, compares adjacent elements, and swaps 
    them if they are in the wrong order. 
    Use Case: Only viable for very small datasets or mostly-sorted data.
    """
    n = len(arr)
    # Track if a swap happened; if a full pass has no swaps, the array is sorted.
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]  # Swap in-place
                swapped = True
        if not swapped:
            break
    return arr


# =========================================================================
# 2. SELECTION SORT (O(n²) - Crane Operator)
# =========================================================================
def selection_sort(arr):
    """
    Divides the array into a sorted and an unsorted region. It repeatedly 
    finds the minimum element from the unsorted region and puts it at the beginning.
    Use Case: Useful when physical memory writes/swaps are highly expensive, 
    as it guarantees a maximum of O(n) swaps.
    """
    n = len(arr)
    for i in range(n):
        min_idx = i
        # Scan the remaining unsorted elements to find the absolute minimum
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
                
        # Swap the found minimum element with the first element of unsorted region
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr


# =========================================================================
# 3. INSERTION SORT (O(n²) - Card Shuffler)
# =========================================================================
def insertion_sort(arr):
    """
    Consumes elements one by one from the unsorted region and inserts them 
    into their correct position relative to the already sorted elements.
    Use Case: Excellent for small datasets (n < 50) or live data streams 
    where new items are continuously appended to an already sorted list.
    """
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        # Move elements of arr[0..i-1] that are greater than key
        # to one position ahead of their current position
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr


# =========================================================================
# 4. MERGE SORT (O(n log n) - Parallel Assembly Line)
# =========================================================================
def merge_sort(arr):
    """
    A Divide and Conquer algorithm. Recursively splits the array in half, 
    sorts the sub-arrays, and merges them back together.
    Use Case: Perfect for External Sorting (sorting datasets too large to 
    fit into RAM) and highly useful in distributed computing frameworks.
    """
    if len(arr) <= 1:
        return arr

    # 1. Divide Phase
    mid = len(arr) // 2
    left_half = merge_sort(arr[:mid])
    right_half = merge_sort(arr[mid:])

    # 2. Conquer/Merge Phase
    return _merge(left_half, right_half)

def _merge(left, right):
    """Helper function to merge two sorted arrays into one."""
    sorted_result = []
    i = j = 0

    # Compare elements from both sub-arrays and pull the smaller one
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            sorted_result.append(left[i])
            i += 1
        else:
            sorted_result.append(right[j])
            j += 1

    # If any elements remain in left or right, append them out-of-the-box
    sorted_result.extend(left[i:])
    sorted_result.extend(right[j:])
    return sorted_result


# =========================================================================
# 5. QUICK SORT (O(n log n) - Pivot Router)
# =========================================================================
def quick_sort(arr):
    """
    A Divide and Conquer algorithm. Chooses a 'pivot' element and partitions 
    the other elements into two sub-arrays (less than vs. greater than pivot).
    Use Case: The standard general-purpose sorting champion. Highly cache-efficient 
    because it operates in-place inside system memory without duplicating arrays.
    """
    if len(arr) <= 1:
        return arr
    else:
        # Using the middle element as the pivot to optimize against pre-sorted data
        pivot = arr[len(arr) // 2]
        left = [x for x in arr if x < pivot]
        middle = [x for x in arr if x == pivot]
        right = [x for x in arr if x > pivot]
        
        return quick_sort(left) + middle + quick_sort(right)
```

```python
# =========================================================================
# ENGINE BENCHMARK EXECUTION SCRIPT
# =========================================================================
if __name__ == "__main__":
    print("=" * 60)
    print("LOGISTICS CONVEYOR BELT SORTING BENCHMARK ENGINE")
    print("=" * 60)

    # Generate a realistic unsorted dataset of 2,500 packages
    DATA_VOLUME = 2500
    master_dataset = [random.randint(1, 10000) for _ in range(DATA_VOLUME)]
    
    print(f"Generated Dataset: {DATA_VOLUME} Unsorted Cargo Packages.")
    print(f"Sample data (First 10 items): {master_dataset[:10]}\n")
    print("Running simulations... Please wait.")
    print("-" * 60)

    # Dictionary containing references to our algorithms
    sorting_algorithms = {
        "1. Bubble Sort (O(n²))"    : bubble_sort,
        "2. Selection Sort (O(n²))" : selection_sort,
        "3. Insertion Sort (O(n²))" : insertion_sort,
        "4. Merge Sort (O(n log n))": merge_sort,
        "5. Quick Sort (O(n log n))": quick_sort,
    }

    # Execute each algorithm on an identical copy of the dataset
    for name, algo_func in sorting_algorithms.items():
        # Create a deep, independent copy of the data to ensure fairness
        dataset_copy = copy.deepcopy(master_dataset)
        
        # Benchmark structural execution time
        start_time = time.perf_counter()
        sorted_output = algo_func(dataset_copy)
        end_time = time.perf_counter()
        
        execution_duration = (end_time - start_time) * 1000 # Convert to milliseconds
        
        # Verify correctness to ensure structural integrity
        is_correctly_sorted = sorted_output == sorted(master_dataset)
        status_token = "PASS" if is_correctly_sorted else "FAIL"

        print(f"{name:<28} | Duration: {execution_duration:>8.2f} ms | Integrity: {status_token}")

    print("-" * 60)
    print("Architectural Takeaway:")
    print("Notice how the O(n log n) algorithms (Merge and Quick Sort) solve")
    print("the entire list in single-digit milliseconds, while the O(n²)")
    print("quadratically slow down even on small pools of 2,500 items.")
    print("=" * 60)
```
