```python
from collections import deque

# ==========================================
# 1. ENTERPRISE STACK IMPLEMENTATION (LIFO)
# ==========================================
class EnterpriseStack:
    def __init__(self):
        # We use a deque container because it is optimized for fast end operations
        self.container = deque()

    def push(self, item):
        """Add an item to the top of the stack. Time Complexity: O(1)"""
        self.container.append(item)

    def pop(self):
        """Remove and return the top item. Time Complexity: O(1)"""
        if self.is_empty():
            raise IndexError("Pop executed on an empty stack.")
        return self.container.pop()

    def peek(self):
        """View the top item without removing it."""
        if self.is_empty():
            return None
        return self.container[-1]

    def is_empty(self):
        return len(self.container) == 0

    def display(self):
        # Displaying top element on the right
        print("Stack (Bottom -> Top):", list(self.container))


# ==========================================
# 2. ENTERPRISE QUEUE IMPLEMENTATION (FIFO)
# ==========================================
class EnterpriseQueue:
    def __init__(self):
        self.container = deque()

    def enqueue(self, item):
        """Add an item to the back of the queue. Time Complexity: O(1)"""
        self.container.append(item)

    def dequeue(self):
        """Remove and return the front item. Time Complexity: O(1)"""
        if self.is_empty():
            raise IndexError("Dequeue executed on an empty queue.")
        # popleft() is a true O(1) operation on a deque
        return self.container.popleft()

    def is_empty(self):
        return len(self.container) == 0

    def display(self):
        # Displaying front element on the left
        print("Queue (Front -> Back):", list(self.container))
```

```python

# === 1. DEMONSTRATING THE STACK (LIFO) ===
print("--- 1. EXECUTING MICROSERVICE CALL STACK ---")
call_stack = EnterpriseStack()

# Simulating service execution flow
print("Service_A executes -> calls Service_B -> calls Service_C")
call_stack.push("Service_A_Frame")
call_stack.push("Service_B_Frame")
call_stack.push("Service_C_Frame")
call_stack.display()

# Simulating resolution
print(f"\nResolving execution. Finished: {call_stack.pop()}")
print(f"Resolving execution. Finished: {call_stack.pop()}")
print("Current state of call stack:")
call_stack.display()
print("-" * 50)


# === 2. DEMONSTRATING THE QUEUE (FIFO) ===
print("\n--- 2. EXECUTING ASYNCHRONOUS VIDEO TRANSCODING QUEUE ---")
video_queue = EnterpriseQueue()

# Ingesting heavy background tasks
print("Ingesting incoming video transcoding jobs...")
video_queue.enqueue("Task_1: Render 4K resolution")
video_queue.enqueue("Task_2: Render 1080p resolution")
video_queue.enqueue("Task_3: Generate Subtitles")
video_queue.display()

# Workers processing jobs
print(f"\nWorker Node 1 picked up: {video_queue.dequeue()}")
print(f"Worker Node 2 picked up: {video_queue.dequeue()}")

print("Current state of pending job queue:")
video_queue.display()
```
