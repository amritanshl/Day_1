Defining the "Clue" (The Node)

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
```

Define the Linked List Wrapper to manage the Nodes

```python
class LinkedList:
    def __init__(self):
        self.head = None
        
    def insert_at_beginning(self, data):
        new_node = Node(data)      # 1. Create a new box
        new_node.next = self.head  # 2. Tell the new box to point to the current start
        self.head = new_node       # 3. Tell the list that the new box is now the start
        
        #O(n)
    def insert_at_end(self, data):
        new_node = Node(data)
        
        # If the list is completely empty, make this the head
        if self.head is None:
            self.head = new_node
            return

        # Otherwise, start at the head and walk to the very last node
        current = self.head
        while current.next is not None:
            current = current.next
        
        # Change the last node's next pointer from None to our new node
        current.next = new_node
    def delete_from_beginning(self):
        if self.head is None:
            print("Error: The list is already empty.")
            return
        
        # Move the head pointer to the second node. 
        # The original first node becomes disconnected.
        self.head = self.head.next
        
    def display_list(self):
        current = self.head
        chain = []
        
        while current is not None:
            chain.append(str(current.data))
            current = current.next
            
        chain.append("None")
        print(" -> ".join(chain))
```

```python
# 1. Initialize an empty version of our Google Docs Undo History
print("Initializing Undo History Buffer...")
undo_history = LinkedList()

# 2. User performs actions. We insert them at the beginning.
# This simulates the newest history state always being right at the front.
print("\nUser typed 'Hello', then highlighted text, then changed color to Red.")
undo_history.insert_at_beginning("Action 1: Typed 'Hello'")
undo_history.insert_at_beginning("Action 2: Highlighted Text")
undo_history.insert_at_beginning("Action 3: Changed Color to Red")

# Show how the chain looks. 
# Notice the most recent action is at the Head.
print("Current History Chain:")
undo_history.display_list()

# 3. User presses Ctrl+Z (Undo)
# We instantly drop the head node without shifting any other data in memory.
print("\nUser hits Ctrl+Z (Undoing the color change)...")
undo_history.delete_from_beginning()

print("Updated History Chain:")
undo_history.display_list()

# 4. Adding a background logging process to the end of the chain
print("\nSystem appends a background autosave marker to the end of the history...")
undo_history.insert_at_end("System: Autosave Point Alpha")

print("Final History Chain:")
undo_history.display_list()
```
